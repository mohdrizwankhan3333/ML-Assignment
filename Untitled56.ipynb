```python
'''Q1. What is an ensemble technique in machine learning?

An ensemble technique combines multiple machine learning models to improve overall performance and predictive accuracy.

Q2. Why are ensemble techniques used in machine learning?

Ensemble techniques reduce overfitting, increase robustness, and enhance the generalization capability of models by leveraging the 
strengths of multiple algorithms.

Q3. What is bagging?

Bagging (Bootstrap Aggregating) is an ensemble method that improves model stability and accuracy by training multiple models on 
different bootstrap samples and averaging their predictions.

Q4. What is boosting?

Boosting is an ensemble technique that sequentially trains models, each focusing on correcting errors made by previous models, and 
combines their outputs to enhance overall performance.

Q5. What are the benefits of using ensemble techniques?

Ensemble techniques enhance prediction accuracy, reduce variance, and improve model robustness, leading to more reliable and stable results.

Q6. Are ensemble techniques always better than individual models?

Not always. Ensemble techniques usually improve performance, but in some cases, simpler individual models may suffice or perform better 
due to lower complexity and computational cost.

Q7. How is the confidence interval calculated using bootstrap?

Bootstrap confidence intervals are calculated by repeatedly resampling the data with replacement, computing the statistic for each 
sample, and determining the interval from the resampled distribution.

Q8. How does bootstrap work and What are the steps involved in bootstrap?

Bootstrap involves: 1) Resampling the dataset with replacement, 2) Calculating the statistic for each resample, 3)
Repeating the process many times, 4) Analyzing the distribution of the statistic.

Q9. A researcher wants to estimate the mean height of a population of trees. They measure the height of a sample of 50 trees 
and obtain a mean height of 15 meters and a standard deviation of 2 meters. Use bootstrap to estimate the 95% confidence interval 
for the population mean height.

1. Resample the data 1000 times with replacement.
2. Calculate the mean for each resample.
3. Determine the 2.5th and 97.5th percentiles of these means.
4. These percentiles form the 95% confidence interval.
'''
import numpy as np

# Original sample data
sample_size = 50
mean_height = 15
std_dev = 2
np.random.seed(0)  # For reproducibility
sample_data = np.random.normal(mean_height, std_dev, sample_size)

# Bootstrap resampling
num_bootstraps = 1000
bootstrap_means = np.zeros(num_bootstraps)
print(bootstrap_means)
for i in range(num_bootstraps):
    bootstrap_sample = np.random.choice(sample_data, size=sample_size, replace=True)
    bootstrap_means[i] = np.mean(bootstrap_sample)

# Calculate 95% confidence interval
confidence_interval = np.percentile(bootstrap_means, [2.5, 97.5])

print("95% Confidence Interval:", confidence_interval)
```
